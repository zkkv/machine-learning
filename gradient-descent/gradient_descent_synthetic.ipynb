# Gradient Descent - Synthetic

This notebook explores the gradient descent algorithm on various synthetic functions.

*Note: variable names are reused, so it's best to run all cells to avoid nonsensical plots.*

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
```

The update rule we utilize in the gradient descent is:
$$
\theta := \theta - \alpha \nabla J(\bold x),
$$

where $J(\bold x)$ is the objective function that depends on vector $\bold x \in \R^n$.

Note that unlike with real objective functions, $J$ does not depend on collected data, so it's not appropriate to talk about "batch" or "stochastic" gradient descent. This is reflected in the `f_prime` functions below that calculate the gradient for a given point `theta` in the parameter space but do not rely on either `X` or `y`.

```python
class Metadata:
	def __init__(self, path, initial_guess, convergence_delta, it, max_iter):
		self.path = path
		self.initial_guess = initial_guess
		self.convergence_delta = convergence_delta
		self.max_iter = max_iter
		self.iter = it

	def __str__(self):
		return f"{self.initial_guess = }\n" \
			+ f"{self.convergence_delta = }\n" \
			+ f"{self.iter = }\n" \
			+ f"{self.max_iter = }"


def gradient_descent(data, gradient_function, learning_rate, initial_guess = None, convergence_delta = 0.00001, max_iter = 10000):
	if initial_guess == None:
		mins = np.min(data, axis=0)
		maxs = np.max(data, axis=0)
		initial_guess = np.atleast_1d(np.random.uniform(mins, maxs))

	initial_guess = np.atleast_1d(initial_guess)
	prev_theta = initial_guess
	theta = initial_guess

	path = []
	path.append(theta)

	n_iter = 0
	while True:
		prev_theta = theta
		theta = prev_theta - learning_rate * gradient_function(prev_theta)
		path.append(theta)
		n_iter += 1

		if np.sum(np.abs(prev_theta - theta)) <= convergence_delta or n_iter >= max_iter:
			break

	return theta, Metadata(np.array(path), initial_guess, convergence_delta, n_iter, max_iter)
```

## Parabola

```python
def f(x):
	return np.pow(x, 2)

def f_prime(point):
	return np.array([2 * point[0]])

y = f(x)
learning_rate = 0.2
minimum, metadata = gradient_descent(x, f_prime, learning_rate)

plot_2D(minimum, metadata, "Parabola")
```

## Rosenbrock function

```python
def f(x, y):
	a = 1
	b = 1
	return (a - x)**2 + b * (y - x**2)**2

def f_prime(point):
	a = 1
	b = 1
	x = point[0]
	y = point[1]

	return np.array([
		2 * (x - a) + 4 * b * x * (x**2 - y), 
		2 * b * (y - x**2)
	])

x  = np.arange(-5, 5, 0.01)
y  = np.arange(-5, 5, 0.01)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

data = np.stack((x, y), axis=1)

learning_rate = 0.05
minimum, metadata = gradient_descent(data, f_prime, learning_rate, initial_guess=[-2, -3], max_iter=1000)

plot_3D(minimum, metadata, "Rosenbrock function", rotation=(60, 30))
```

## Himmelblau's function

```python
clip_value = 1_000_000

def f(x, y):
	result = (x**2 + y - 11)**2 + (x + y**2 - 7)**2 
	return np.clip(result, -clip_value, clip_value)

def f_prime(point):
	x = point[0]
	y = point[1]

	result = np.array([
		2 * (2 * x * (x**2 + y - 11) + x + y**2 - 7), 
		2 * (x**2 + 2 * y * (x + y**2 - 7) + y - 11)
	])

	return np.clip(result, -clip_value, clip_value)

x  = np.arange(-5, 5, 0.01)
y  = np.arange(-5, 5, 0.01)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

data = np.stack((x, y), axis=1)

learning_rate = 0.01
minimum, metadata = gradient_descent(data, f_prime, learning_rate, initial_guess=[0, 0], max_iter=1000)

plot_3D(minimum, metadata, "Himmelblau's function", rotation=(285, 40))
```

## Multidimensional Rosenbrock function
